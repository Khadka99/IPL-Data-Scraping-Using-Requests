```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive'}

current_page = 1

data = []

max_page = 512

while current_page <= max_page:
    print(f"Printing Page No:-{current_page}")

    url = f"https://stats.espncricinfo.com/ci/engine/stats/index.html?class=6;filter=advanced;orderby=batted_score;page={current_page};runsmax1=200;runsval1=runs;team=4340;team=4341;team=4342;team=4343;team=4344;team=4345;team=4346;team=4347;team=4787;team=4788;team=5143;team=5843;team=5845;team=6903;team=6904;template=results;type=batting;view=innings"

    page = requests.get(url,headers=headers)

    soup = BeautifulSoup(page.text, "html.parser")
    
    lists = soup.find_all("tr",class_=["data2","note"])
    
    for list_item in lists:
                item = {}
            
                # Extracting the player's name
                # name_tag = list_item.find("a", class_="data-link")
                # item['Name'] = name_tag.text.strip() if name_tag else "No name available"
                fields = list_item.find_all("td")
                item['Date'] = fields[11].text.strip() if len(fields) > 11 else "No Date information"
                # Extract player's name
                name_tag = fields[0].find("a", class_="data-link")
                item['Name'] = name_tag.text.strip() if name_tag else "No name available"
        
                team_row = list_item.find_next_sibling('tr', class_='note')
                if team_row:
                    team_name = team_row.text.strip().strip('()')
                
                if len(fields) >= 11:
                    item['Runs'] = fields[1].text.strip() if len(fields) > 1 else "No runs available"
                    item['Mins'] = fields[2].text.strip()
                    item['Ball_faced'] = fields[3].text.strip()
                    item['4s'] = fields[4].text.strip()
                    item['6s'] = fields[5].text.strip()
                    item['SR'] = fields[6].text.strip()
                    item['Innings'] = fields[7].text.strip()
                    item['Opposition'] = fields[9].text.strip() if len(fields) > 9 else "No opposition available"
                    item['Ground'] = fields[10].text.strip() if len(fields) > 10 else "No ground available"
                    item['Team'] = team_name
                    data.append(item)

    current_page += 1
```
